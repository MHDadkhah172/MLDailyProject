```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/car data.csv')
df = pd.DataFrame(data)
df = df.drop('Car_Name', axis=1)
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.regplot(x='Year', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Year')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Present_Price', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Present_Price')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Kms_Driven', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Kms_Driven')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Fuel_Type', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Fuel_Type')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Seller_Type', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Seller_Type')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Transmission', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Transmission')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Owner', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Owner')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```
