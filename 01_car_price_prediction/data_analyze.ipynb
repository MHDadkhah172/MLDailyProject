```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('data/car data.csv')
df = pd.DataFrame(data)
df.drop('Car_Name', axis=1, inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.regplot(x='Year', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Year')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Present_Price', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Present_Price')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
df = df[df['Present_Price'] < 80]
```

```python
sns.regplot(x='Kms_Driven', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Kms_Driven')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Fuel_Type', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Fuel_Type')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Seller_Type', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Seller_Type')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Transmission', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Transmission')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='Owner', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Owner')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
##### deleting one row that has 3 Owner
df = df[df['Owner'] < 3]
```

```python
plt.scatter(df['Year'], df['Kms_Driven'], alpha=0.5)
plt.xlabel('Year')
plt.ylabel('Kms_Driven')
plt.grid(True)
plt.show()
```

```python
max_kms_index = df['Kms_Driven'].idxmax()
df.loc[max_kms_index]
```

```python
###### delete row with 500k km Kms_Driven
df = df[df['Kms_Driven'] < 400000]
```

```python
plt.scatter(df['Fuel_Type'], df['Present_Price'], alpha=0.5)
plt.xlabel('Fuel_Type')
plt.ylabel('Present_Price')
plt.grid(True)
plt.show()
```

```python
max_Presint_Price = df['Present_Price'].idxmax()
df.loc[max_Presint_Price]
```

```python
plt.scatter(df['Year'], df['Selling_Price'], alpha=0.5)
plt.xlabel('Year')
```

```python
df.describe()
```

```python
#### creating new dataset with less noise
df.to_csv('data/dataset_logical_filter.csv', index=False)
```

```python
sns.regplot(x='Present_Price', y='Selling_Price', data=df, scatter_kws={'alpha':0.5})
plt.xlabel('Present_Price')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Present_Price'], df['Selling_Price'], alpha=0.5)
plt.xlabel('Present_Price')
plt.ylabel('Selling_Price')
plt.grid(True)
plt.axline((0, 0), slope=1, color='black', linestyle='--')
max_val = df['Present_Price'].max()
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='y = x (No Depreciation)')
min_ratio = (df['Selling_Price'] / df['Present_Price']).min()
x_vals = np.array([0, max_val])
y_vals = min_ratio * x_vals
plt.plot(x_vals, y_vals,color="purple", label=f'Market Floor (y = {min_ratio:.2f}x)')
print(min_ratio)
plt.show()
```
